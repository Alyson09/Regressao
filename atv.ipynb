```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nf = pd.read_csv ('./Data/aluguel_uma_variavel_com_ruido.csv')
```

```python
nf.head()
```

```python
nf.describe()
```

```python
plt.scatter(nf['Area'], nf['Preco'])
plt.xlabel('Área (m²)')
plt.ylabel('Preço do Aluguel (R$)')
plt.title('Preço do Aluguel x Área da Casa')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
t = nf[['Area']]
tr = nf ['Preco']

#Dividindo entre Treino/Teste
t_train, t_test, tr_train, tr_test = train_test_split(t,tr, test_size = 0.3, random_state=42)
```

```python
#Treinando o modelo
modelo = LinearRegression()
modelo.fit(t_train, tr_train)

#Prevendo o erro
tr_pred = modelo.predict(t_test)

#imprimindo a avaliação
print(f"R²: {r2_score(tr_test, tr_pred):.2f}")
```

```python
plt.scatter(nf['Area'], nf['Preco'], color = 'red')
plt.plot(t_test, tr_pred, color='black')
plt.xlabel('Área (m²)')
plt.ylabel('Preço do Aluguel (R$)')
plt.title('Preço do Aluguel x Área da Casa')
plt.show()
```
